Все, что нужно будет делать — запускать методы из `sklearn`. Вам предлагается выяснить, какое распределение лучше использовать в наивном байесовском классификаторе в зависимости от вида признаков.

Загрузите датасеты `digits` и `breast_cancer` из `sklearn.datasets`. Выведите несколько строчек из обучающих выборок и посмотрите на признаки. С помощью `sklearn.cross_validation.cross_val_score` c настройками по умолчанию и вызова метода `mean()` у возвращаемого этой функцией `numpy.ndarray`, сравните качество работы наивных байесовских классификаторов на этих двух датасетах. Для сравнения предлагается использовать `BernoulliNB`, `MultinomialNB` и `GaussianNB`. Насколько полученные результаты согласуются с рекомендациями из лекций?

Два датасета, конечно, еще не повод делать далеко идущие выводы, но при желании вы можете продолжить исследование на других выборках (например, из `UCI` репозитория).

```python
from sklearn import datasets, model_selection
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
import numpy as np
import pandas as pd
```

```python
digits = datasets.load_digits()
breast_cancer = datasets.load_breast_cancer()
```

```python
digits_df = pd.DataFrame(digits.data)
digits_df['target'] = digits.target

digits_df.head()
```

```python
cancer_df = pd.DataFrame(breast_cancer.data)
cancer_df.columns = breast_cancer.feature_names
cancer_df['target'] = breast_cancer.target

cancer_df.head()
```

```python
def write_answer(ans, n):
    with open("ans{}.txt".format(n), "w") as fout:
        fout.write(str(ans))
```

```python
digits_scoring = []
cancer_scoring = []

for classifier in [BernoulliNB(), MultinomialNB(), GaussianNB()]:
    digits_scoring.append(model_selection.cross_val_score(classifier, digits.data, digits.target).mean())
    cancer_scoring.append(model_selection.cross_val_score(classifier, breast_cancer.data, breast_cancer.target).mean())

print(digits_scoring)
print(cancer_scoring)
```

### Вопрос 1
Каким получилось максимальное качество классификации на датасете breast_cancer?

```python
write_answer(max(cancer_scoring), 1)
```

### Вопрос 2
Каким получилось максимальное качество классификации на датасете digits?

```python
write_answer(max(digits_scoring), 2)
```
